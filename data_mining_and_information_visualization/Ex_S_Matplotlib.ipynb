## (1) 折れ線グラフ / Line graph
線の太さを変えてみよう。<br>
Please change the thickness of the line.<br>

````python
#グラフを業内に表示
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,10,100)
y = np.exp(-x)
plt.plot(x,y)
plt.title('Line graph')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

## (2) 散布図 / Scatter plot
マークの大きさを変えてみよう。/ Please change the size of the markers.<br>
軸のラベルのフォントサイズを変えてみよう。/ Please change the font size of the axis labels.<br>

````python
#グラフを業内に表示
%matplotlib inline
import matplotlib.pyplot as plt
from numpy.random import rand

x = rand(100)
y = rand(100)
plt.scatter(x,y)
plt.title('Scattered plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
````

## (4) 円グラフ / Pie charts

````python
import numpy as np
import matplotlib.pyplot as plt

# 参考サイト　http://一人暮らし.tokyo/living-cost/
x = np.array([48000, 11000, 33000, 6500, 5000, 5000])
plt.pie(x)
label = ["Housing", "Utility", "Food", "Phone", "Misc", "Clothings"]
plt.pie(x, labels=label)
plt.title('Average living costs per month of undergraduate students in Fukuoka')
# 日本語化ライブラリをインストールしてあれば、日本語でも書ける。
# plt.title('福岡の大学生の平均的な生活費内訳')
plt.show()
````

## (5) バブルチャート / Bubble charts

````python
# libraries
import matplotlib.pyplot as plt
import numpy as np
#グラフを業内に表示
%matplotlib inline

# number of data
n=40

# create data
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)
 
# use the scatter function
plt.scatter(x, y, s=z*1000, alpha=0.5)
plt.show()
````

## (6) ヒストグラム / Histogram
色を変えてみよう<br>
ヒストグラムのビンの数を変えてみよう。<br>
参考サイト  http://hxn.blog.jp/archives/8930480.html

````python
#グラフを業内に表示
%matplotlib inline
import matplotlib.pyplot as plt
from numpy.random import normal,rand

x = normal(size=200) # 正規分布で乱数を200個生成する。
plt.hist(x,bins=30) # ビンの数を変えるにはbinsを変える。
plt.title('Histogram') # グラフのタイトル
plt.xlabel('x') # x軸のラベル
plt.ylabel('Frequency') # y軸のラベル
plt.show()
````

## (7) 3Dグラフ / 3D plot

````python
#グラフを業内に表示
%matplotlib inline
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.coolwarm)
plt.show()
````

## 複数のグラフを並べてプロットするには??

参考サイト: [matplotlibのグラフを入れ子に！subplotを使おう！](https://www.sejuku.net/blog/61011)


````python
x = np.linspace(0, 5*np.pi, 100)
 
y_sin = np.sin(x)
y_cos = np.cos(x)

# たてに二つのパネルに並べる。
# １つ目のプロット
plt.subplot(2,1,1)
plt.plot(x, y_sin, label="sin")
plt.legend()
 
# 2つ目のプロット
plt.subplot(2,1,2)
plt.plot(x, y_cos, label="cos")
plt.legend()

# よこに二つのパネルに並べてみよう。
````

````python
x = np.linspace(0, 5*np.pi, 100)
 
y_sin = np.sin(x)
y_cos = np.cos(x)

# 2x2 のパネルに並べる。
# １つ目のプロット
plt.subplot(2,2,1)
plt.plot(x, y_sin, label="sin")
plt.legend()
 
# 2つ目のプロット
plt.subplot(2,2,2)
plt.plot(x, y_cos, label="cos")
plt.legend()
 
# 3つ目のプロット
plt.subplot(2,2,3)
y_tan = np.tan(x)
plt.plot(x, y_tan, label="tan")
plt.legend()
 
# 4つ目のプロット
plt.subplot(2,2,4)
y_random = np.random.randn(len(y_tan))
plt.plot(x, y_random, label="random")
plt.legend()
````
