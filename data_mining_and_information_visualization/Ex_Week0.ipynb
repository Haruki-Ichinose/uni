### セルの種類 : Code / Markdown
* Pythonで実行したい内容は、セルの種類を　Code　にしてください。<br>
ショートカットキー「Esc-Y」 でCodeモードになります。<br>
![ipynb_window](img/w0_ipynb_cell_code.jpg)<br>
セルの種類が Code の時は、以下のように書くと、シャープマークより後ろがコメントアウトされます。


```
# 何を書いてもプログラミングコードの内容に影響しません。
# コメントとして、プログラムの説明を書いておくと、後から見直したときにわかりやすくなります。
# 人とプログラムを共有するときも、作業がしやすくなります。
```


* セルの種類が　Markdown の時は文章を入力できます。<br>
ショートカットキー「Esc-M」 でMarkdownモードになります。<br>
![ipynb_window](img/w0_ipynb_cell_markdown.jpg) <br>
MarkdownはHTMLで文章を制御できます。

### 代表的なショートカットキー
|  Action  |  Shortcut keys  |
| ---- | ---- |
|  セルの実行 / Run selected cells and don't advance |  Ctrl-Enter  |
|  セルの実行して下のセルに移動 / Run selected cells and advance |  Shift-Enter  |
|  新しいセルを下に挿入 / Insert a cell below |  Esc-B  |
|  新しいセルを上に挿入 / Insert a cell above |  Esc-A  |
|  セルの種類を　Code にする / Set the cell type to Code |  Esc-Y  |
|  セルの種類を　Markdown にする / Set the cell type to Markdown |  Esc-M  |
|  選択したセルを削除 / Remove cells |  Esc-DD  |

ちなみに、Escボタンは遠い(キーボードの左上の一番隅)ので、「Ctrl-M」でもEscボタンと同じ操作になります。<br>

```
<h1># 大きい</h1>
<h2>## 大きい</h2>
<h3>### 大きい</h3>
```
数字が小さいほど大きな文字になります。<br>
数字の大きさと#の数が同じときは、同じ大きさの文字が表示されます。

## 改行
`<br>`で改行します。

## リンクの貼り方
`[こちら](https://www.kyushu-u.ac.jp/ja/)`と書くとこうなります。<br>
[こちら](https://www.kyushu-u.ac.jp/ja/)

ローカルにファイルもリンクできます。<br>
`[こちらをクリック](Report_Week1.ipynb)`と書くとこうなります。<br>
[こちらをクリック](Report_Week1.ipynb)

## 画像の貼り方
以下のように書くと、画像も貼れます。<br>
画像を別のフォルダに入れてる場合には、画像へのパスも含めて書きます。<br>

``` shell
![代替テキスト](画像のファイル名)
![ipynb_window](img/ipynb_cell_markdown.jpg)
```
